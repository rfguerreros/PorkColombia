```python
#Librerías
from pathlib import Path
import pandas as pd
import os
import numpy as np
```

```python
#Funciones
def ls (path):
    return [obj.name for obj in Path(path).iterdir() if obj.is_file()]
#--------------------------------------------------------------------------------------------------------------------#
def clean(df, col):
    if col != None:
        a = df.iloc[col[0]:col[1], col[2]:col[3]]#df.pop('Unnamed: '+str(c))
        a.columns = df.iloc[col[0]-1, col[2]:col[3]]
        a.dropna(how='all', inplace=True)
        return a
#--------------------------------------------------------------------------------------------------------------------#
def cleaner(df):
    try:
        head = df[df[df.columns[1]]=='DEPARTAMENTO'].index[0]
        #print(head)
        tmp = list(map (lambda x: isinstance(x,str), list(df.iloc[head])))
        for j in range(len(tmp)-1):
            if tmp[j] == False and tmp[j+1] == False: break;

        #s = pd.Series(list(df.iloc[head].isna()))
        aux=[head+1, df[df[df.columns[1]].isna()].index[-1],1, j]
        return aux
    except IndexError:
        print('Tabla sin enfermedad')
        return None
#--------------------------------------------------------------------------------------------------------------------#
```

```python
path='BBDD'#Sólo guardar archivos .xlsx
files=ls(path)
df = []
#-----Lectura de archivos y creación de DF por hojas --------#
for f in files:
    sheets = pd.ExcelFile(open(path+'/'+f, 'rb'))
    for s in sheets.sheet_names:
        df.append(pd.read_excel(open(path+'/'+f, 'rb'), sheet_name=s))
#---------Clean data-----------#
data = pd.Series(map(cleaner, df))
tables = list(map(clean,df,data))
#-----------Tablas extraídas----#
tables[2]
```
