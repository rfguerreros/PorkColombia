```python
file = '12 INFORMA DIC 2015 PRONAVICOLA.xls'
filename = file.replace('.xls','')
filename = filename.replace(' ','')
book = xlrd.open_workbook(file)


begi_dict = {}
for sheet in book.sheets():
    begi_dict[sheet.name] = []
    for row in range(sheet.nrows):
        deptobool = False
        interbool = False
        for value in sheet.row_values(row):
            if 'DEPARTAMENTO' in str(value):
                deptobool = True
            if 'GRANJA' in str(value):
                interbool = True
        if deptobool == True and interbool == True:
            begi_dict[sheet.name].append(row)

end_dict = {}
for sheet in book.sheets():
    end_dict[sheet.name] = []
    if len(begi_dict[sheet.name])==0:
        continue
    for headrow in begi_dict[sheet.name]:
        looking = True
        found = False
        for row in range(headrow+1,sheet.nrows):
            for value in sheet.row_values(row):
                if 'TOTAL MES' in str(value):
                    end_dict[sheet.name].append(row)
                    found = True
            if found == True:
                break
        if found == False:
            print(sheet.name)
            print('ALERTTTTTTTTTTTTTTTTTTTTTT\n')
    for endrow in range(len(end_dict[sheet.name])-1):
        if end_dict[sheet.name][endrow] == end_dict[sheet.name][endrow+1]:
            print(sheet.name)
            print('ALERTTTTTTTTTTTTTTTTTTTTTT\n')

for i in begi_dict:
    if len(begi_dict[i]) == len(end_dict[i]):
        None
    else:
        print(i,'MAAAAAAL')
```

```python
for sheet in book.sheets():
    print(sheet.name)
    if len(begi_dict[sheet.name])==0:
        continue
    firstcolumn = 0
    for value in sheet.row_values(begi_dict[sheet.name][0]):
        if value!='':
            break
        firstcolumn = firstcolumn + 1
    n_headrow = 0
    df = pd.DataFrame()
    for headrow in begi_dict[sheet.name]:
        endrow = end_dict[sheet.name][n_headrow]
        if endrow == headrow + 1:
            n_headrow = n_headrow + 1
            print(sheet.name,'NO HAY NADA')
            continue
        checker = False
        checkercounter = 0
        for value in sheet.row_values(headrow+1):
            if '' == value:
                checkercounter = checkercounter + 1
        if endrow == headrow + 2 and checkercounter> 2:
            n_headrow = n_headrow + 1
            print(sheet.name,'NO HAY NADA')
            continue
        dftemp = pd.read_excel(file,sheet_name=sheet.name,dtype=type(''),
                     header=headrow,nrows=(endrow-headrow)-1)
        if n_headrow == 0:
            headercolumns = dftemp.columns
        if n_headrow > 0 and len(dftemp.columns)==len(headercolumns):
            dftemp.columns = headercolumns
        df = df.append(dftemp)
        n_headrow = n_headrow + 1
    if df.empty:
        continue
    df.to_csv(filename+'_'+ sheet.name +'.csv', index=False, encoding='utf-8')
```
